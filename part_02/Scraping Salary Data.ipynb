```python
def get_team_table(url):
    res = requests.get(url)
    cap_soup = BeautifulSoup(res.content, 'lxml')
    team_name = cap_soup.find('div', {'class': 'team-name'}).find('h1').text
    table = cap_soup.find('div', {'class': 'teams'}).find('tbody')
    team_salary = []
    for row in table.find_all('tr'):
        team = {}
        for element in row:
            team['Player'] = row.find('a').text
            team['Position'] = row.find('span', {'class': 'cap'}).text
            team['Age'] = row.find('td', {'class': ' center xs-hide'}).text
            team['Base Salary'] = row.find('span', {'class': 'cap info'}).text
            team['Signing Bonus'] = row.find_all('span', {'class': 'cap info'})[1].text
            team['Performance Bonus'] = row.find_all('span', {'class': 'cap info'})[2].text
            team['Total Salary'] = row.find_all('span', {'class': 'cap info'})[3].text
            team['Total Cap Hit'] = row.find_all('span', {'class': 'cap info'})[4].text
            team['Adjusted Cap Hit'] = row.find_all('span', {'class': 'cap info'})[5].text
            team['Cap Percentage'] = row.find('td', {'class': ' center '}).text
            team['Team'] = team_name
        team_salary.append(team)
    return team_salary
```

```python
team_year = get_team_table('https://www.spotrac.com/nhl/anaheim-ducks/cap/2018')
```

```python
teams = team_links
years = ['2011']
year_df = pd.DataFrame()
for team in teams:
    try:
        for year in years:
            url = team + year
            team_year = get_team_table(url)
            team_df = pd.DataFrame(team_year)
            year_df = pd.concat([year_df, team_df], axis=0)
            year_df.reset_index(drop=True, inplace=True)
            time.sleep(3)
    except:
        continue
```
